```python
from datetime import datetime
def convert_date(ds):
    date_object = datetime.strptime(ds, "%b%y")
    return date_object

crime_data["datestamp"] = crime_data['Month Year'].apply(convert_date)
crime_data.set_index("datestamp", inplace=True)
crime_data.sample(5)
```

```python
def crime_graph(age="Adult", sex="Female", start_date="01-01-2000", end_date="01-01-2023", column="Traffic and Related Offences"):
    crime_data[
        (crime_data.Age == age) & 
        (crime_data.Sex == sex) & 
        (crime_data.index < end_date) & 
        (crime_data.index > start_date)
    ][column].plot(title=column)
```

```python

crime_graph()
crime_graph(sex="Male") 



```
